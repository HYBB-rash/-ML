```python
import pandas as pd
```

```python
dataTrain = pd.read_csv('E:/code/ml/lee-ml-code/-ML/hw1/ml2020spring-hw1/train.csv',encoding='utf-8')
dataTrain[0:36][:]
```

```python
dataTrain = dataTrain[dataTrain['測項'].str.contains("PM2.5")]
dataTrain
```

```python
dataTrain = dataTrain.reset_index()
dataTrain = dataTrain.drop(columns='測站')
dataTrain = dataTrain.drop(columns='測項')
dataTrain = dataTrain.drop(columns='index')
```

```python
dataTrain
```

```python
for i in range(24):
    dataTrain[str(i)] = dataTrain[str(i)].astype(float)
```

```python
list = []
for i in range(24):
    list.append(str(i))
```

```python
dt = dataTrain[list].sum(axis=1)
dt = dt.to_frame()
dt = dt.T
```

```python
dt.rename(columns = {'0':'PM2.5'}, inplace = True)
dataTrain['PM2.5'] = dt['PM2.5']
```

```python
for i in range(24):
    dataTrain = dataTrain.drop(columns=str(i))
dataTrain
```
