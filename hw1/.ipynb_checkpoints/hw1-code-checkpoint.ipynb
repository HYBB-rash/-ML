```python
x = dataTrain['日期']
x = x.values
y = dataTrain['PM2.5']
y = y.values
```

```python
def fx(w, b, x):
    y = np.linspace(b, b, x.shape[0])
    for i in range(w.size):
        y += w[i] * x
        x = x * x
    return y
```

```python
def loss(y, yCap):
    yCap = yCap - y
    yCap = yCap * yCap;
    return np.sum(yCap)
```

```python
def epcho(x, y, yCap, w, b, a):
    for i in range(w.size):
        dw = np.sum((-2 * (y - yCap) * x))
        w[i] -= a * dw;
        x = x * x
    b -= np.sum((2 * (y - yCap)))
    return w, b
```

```python
w = np.zeros(300)
b = 2
```

```python
yCap = fx(w, b, x)
print(loss(y, yCap))
w, b = epcho(x, y, yCap, w, b, 0.1)
w, b
```
